```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t=np.array([1,2,3,4,5])
plt.plot(t**2,color="red")
plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend() # legends with out label
plt.show
```

```python
t=np.array([1,2,3,4,5])
plt.plot(t**2,color="red")
plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend(["squares","cubes"]) # legends with label
plt.show
```

```python
t=np.array([1,2,3,4,5])
line1,=plt.plot(t**2,color="red")
line2,=plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend([line1,line2,],["squares","cubes"]) # legends with labels and handles
plt.show
```

```python
t=np.array([1,2,3,4,5])
line1,=plt.plot(t**2,color="red")
line2,=plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend([line1,line2,],["squares","cubes"],ncol=2,loc="upper right") # ncol used for columns.we can use a numerical value or string for loc.
plt.show
```

```python
t=np.array([1,2,3,4,5])
line1,=plt.plot(t**2,color="red")
line2,=plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend([line1,line2,],["squares","cubes"],ncol=2,loc=6) # ncol used for columns.
plt.show
```

```python
t=np.array([1,2,3,4,5])
line1,=plt.plot(t**2,color="red")
line2,=plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend((0.5,0.5),loc=4 )
plt.show()
```

```python
t=np.array([1,2,3,4,5])
line1,=plt.plot(t**2,color="red")
line2,=plt.plot(t**3,color="black")
plt.xlabel("x-label")
plt.ylabel("y-label" )
plt.title("First plot")
plt.legend(loc="center",bbox_to_anchor=(0.5,0.5,0.5,0.5) )
plt.show()
```
