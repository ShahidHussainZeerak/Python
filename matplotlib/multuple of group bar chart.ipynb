```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["CSE","ISE","ECE","EEE","MECH"]
boys=[10,40,50,30,20]
girls=[200,70,40,80,120]
w=0.3
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
plt.bar(bar1,w,label="boys" ,color="green",)
plt.bar(bar2,w,label="girls", color="blue")
plt.xlabel("course")
plt.ylabel("students")
plt.title("Course vs Students")
plt.xticks(bar1+w/2,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import  numpy as np
```

```python
w=0.3
x=["usa","russia","china","germany","france"]
gold=[20,10,6,24,30]
silver=[100,50,30,120,150]
bronze=[200,100,60,240,300]
bar1=np.arange(len(x))              # locatiuon for each bar.
bar2=[i+w for i in  bar1]
bar3=[i+w for i in  bar2]
plt.bar(bar1,w,label="gold")
plt.bar(bar2,w,label="silver")
plt.bar(bar3,w,label="bronze")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.legend()

plt.show()
```
