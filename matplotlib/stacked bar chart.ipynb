```python
import matplotlib.pyplot as plt
```

```python
course=["CSE","ISE","ECE","EEE","MECH"]
boys=[300,200,400,250,400]
girls=[150,100,200,150,250]
w=0.4
plt.bar(course,boys,w,label="boys" ,color="green")
plt.bar(course,girls,w,label="girls", color="blue")
plt.xlabel("course")
plt.ylabel("Students")
plt.title("Course vs Students")
plt.legend()
plt.show()
```

```python
# we can also create bar chart with error bar.
```

```python
course=["CSE","ISE","ECE","EEE","MECH"]
boys=[30,20,40,50,45]
girls=[15,10,20,25,30]
w=1
plt.bar(course,boys,w,label="boys" ,color="red",yerr=boys)
plt.bar(course,girls,w,label="girls", color="black",xerr=girls)
plt.xlabel("course")
plt.ylabel("Students")
plt.title("Course vs Students")
plt.legend()
plt.show()
```
