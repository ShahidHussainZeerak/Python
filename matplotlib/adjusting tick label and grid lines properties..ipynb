```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4,5])
plt.plot(t**2,t,color="red")
plt.plot(t**3,t,color="orange")
plt.xlim(0,10)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("My first plot")
# Grid lines properties
plt.grid(color="red",which="major",linestyle=":")
plt.tick_params(labelsize=20,labelcolor="red",colors="black",width=10,labeltop=False,labelbottom=True,labelleft=True,labelright=False,labelrotation=120)
plt.show()
```
