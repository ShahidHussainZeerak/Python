```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
t=np.array([1,2,3,4,5])
plt.plot(t**2,t,color="red")
plt.plot(t**3,t,color="orange")
plt.xlim(0,10)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("My first plot")
plt.minorticks_on()
plt.xticks([2,6],["A","B"],color="Black",size="xx-large") # xticks
plt.yticks([2,6],["c","d"],color="red",size="xx-large")# yticks.
plt.show()
```

```python
#length,width,color,size,pad,direction,rotation:
import numpy as np
t=np.array([1,2,3,4,5])
plt.plot(t**2,t,color="red")
plt.plot(t**3,t,color="orange")
plt.xlim(0,10)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("My first plot")
plt.grid()
plt.tick_params(axis="both",color="red",size=10,width=5,pad=10,rotation=90,direction="in",which="minor",right=True,top=True,bottom=False,left=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4,5])
plt.plot(t**2,t,color="red")
plt.plot(t**3,t,color="orange")
plt.xlim(0,10)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("My first plot")
plt.grid()
plt.tick_params(labelsize=20,labelcolor="red")
plt.show()
```
