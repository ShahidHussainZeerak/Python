```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40],[1,2,3,4])
plt.plot([15,25,30,45],[1,2,3,4])
plt.xlabel("x-label",color="red")
plt.ylabel("y-label" ,color="green")
plt.title("First plot")
plt.minorticks_on()
plt.grid(b=True,which="major",axis="both",color="red",linewidth=5)
plt.grid(b=True,which="minor",axis="both",color="blue")
plt.show()
```
