```python
import matplotlib.pyplot as plt
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red"},size=15,labelpad=20)
plt.ylabel("y-label",color="black",size=15,labelpad=20)
plt.title("First plot",{"color":"red"},size=25,backgroundcolor="yellow",loc="left") # loc default value is centre
plt.show()
```
