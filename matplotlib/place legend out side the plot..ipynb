```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4,5])
plt.plot(a**2,a,color="red")
plt.plot(a**3,a,color='green')
plt.xlabel("x-label",)
plt.ylabel("y-label")
plt.title("First plot")
#plt.legend(["squares",'cubes'],loc=(1,1),fontsize="xx-large") # use of loc
plt.legend(["squares",'cubes'],loc="upper left",bbox_to_anchor=(0,0),fontsize="xx-large")# use of bbox_to_anchor
plt.tight_layout()
plt.show()
```

```python
a=np.array([1,2,3,4,5])
plt.plot(a**2,a,color="red")
plt.plot(a**3,a,color='green')
plt.xlabel("x-label",)
plt.ylabel("y-label")
plt.title("First plot")
plt.grid()
plt.legend(["squares",'cubes'],loc="upper left",fontsize="xx-large",facecolor="yellow",edgecolor="red",framealpha=1)# use of bbox_to_anchor
plt.tight_layout()
plt.show()
```

```python
a=np.array([1,2,3,4,5])
plt.plot(a**2,a,color="red")
plt.plot(a**3,a,color='green')
plt.xlabel("x-label",)
plt.ylabel("y-label")
plt.title("First plot")
plt.grid()
#plt.legend(["label1","label2"],markerfirst=False)# markerfirst
#plt.legend(["label1","label2"],shadow=False ,fontsize=20)# shadow
#plt.legend(["label1","label2"],loc="center",frameon=True ,fontsize=20)
plt.legend(["label1","label2"],shadow=False ,fontsize=20,ncol=2,columnspacing=3)
plt.tight_layout()
plt.show()
```

```python
a=np.array([1,2,3,4,5])
l1=plt.plot(a**2,a,color="red")
l2=plt.plot(a**3,a,color='green')
l3=plt.plot(a,a**2,color="red")
l4=plt.plot(a,a**3,color='green')
plt.xlabel("x-label",)
plt.ylabel("y-label")
plt.title("First plot")
plt.grid()
plt.legend(["label1","label2","label3","label4"],title="My label",borderpad=4,labelspacing=3)
plt.tight_layout()
plt.show()
```
