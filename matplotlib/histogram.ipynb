```python
import matplotlib.pyplot as plt 
```

```python
x=[1,12,13,4,5,6,7,8,19,10,9,11,15,14,17,18,19,20,25,24,26,23,33,22,21,40,35,44,47,50]
plt.hist(x,bins=3,ec="red") # bins is used int type
plt.xlabel("age")
plt.ylabel("count")
plt.title("population age count")
plt.show()
```
