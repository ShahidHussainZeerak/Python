```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40,50])
plt.text(2.5,30,"point",bbox={"fc":"yellow","ec":"red","boxstyle":"circle","pad":4})
plt.show()
```

```python
plt.plot([10,20,30,40,50])
plt.annotate("(1,20)",(1,20),(0.5,25),bbox={"fc":"gray","ec":"red","boxstyle":"circle","pad":2})
plt.show()
```
