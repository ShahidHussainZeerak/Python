```python
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,6,8,9,10],[1,2,3,4,5,6,7],"k:") # k=black color and ":,--,-" is line style
plt.plot([5,6,7,8,9,10,11],[1,2,3,4,5,6,7],"c--") # c=cyan color
plt.plot([7,8,9,10,11,12,14],[2,4,6,8,10,12,13],"m-")# m=magenta
plt.xlabel("x label")
plt.ylabel("y label")
plt.title('first plot')
plt.show()
```

```python
plt.plot([5,6,7,8,9,10,11],[1,2,3,4,5,6,7],"b--o") # o circle marker symbol
plt.plot([7,8,9,10,11,12,14],[2,4,6,8,10,12,13],"r->") # > is triangle up marker.
plt.xlabel("x label")
plt.ylabel("y label")
plt.title('first plot')
plt.show()
```
