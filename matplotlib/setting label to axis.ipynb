```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red"},size=15,labelpad=20,)
plt.ylabel("y-label",color="black",size=15,labelpad=20)
plt.title("First plot")
plt.show()
```

```python
# LECTURE :8...text properties of xlabel and y-label function.
```

```python
# 1:fontsize/size: we can use both numerical value and  string values.
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red"},size=15,labelpad=-100,)
plt.ylabel("y-label",color="black",fontsize="x-large",labelpad=20)
plt.show()
```

```python
#2:color:we can change color of text using this properties.
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"orange"},size=25,labelpad=-100,)
plt.ylabel("y-label",color="cyan",fontsize="xx-large",labelpad=20)
plt.show()
```

```python
#3:Alpha: used for transferency..its value is from 0-1...it zero the line  is completely invisible.
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red"},size=15, alpha=1,labelpad=10,)
plt.ylabel("y-label",color="black",fontsize="x-large", alpha=0,labelpad=20)
plt.show()
```

```python
# 4:Background color: change background color of a text.
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red","backgroundcolor":"yellow"},size=15,labelpad=-100,)
plt.ylabel("y-label",color="black",fontsize="x-large",labelpad=20,backgroundcolor="yellow")
plt.show()
```

```python
#5:font style/style:used to change font style of a text.
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red","backgroundcolor":"yellow"},size=15,labelpad=-100, fontstyle="italic")
plt.ylabel("y-label",color="black",fontsize="x-large",labelpad=20,backgroundcolor="yellow",style="oblique")
plt.show()
```

```python
#6:font weight/weight:using this we can bold or light the label...used numerical value from 0-1000 and also a string values.
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red","backgroundcolor":"yellow","weight":"black"},size=15,labelpad=20,)
plt.ylabel("y-label",color="black",fontsize="x-large",labelpad=20,backgroundcolor="yellow",weight=1000)
plt.show()
```

```python
#7:font:
plt.plot([2,4,6,8,10,12],[1,2,3,4,5,6])
plt.xlabel("x-label",{"color":"red","backgroundcolor":"yellow","font":"serif"},size=15,labelpad=-100,)
plt.ylabel("y-label",color="black",fontsize="xx-large",labelpad=20,backgroundcolor="yellow",font="cursive")
plt.show()
```
