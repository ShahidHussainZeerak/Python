```python
import matplotlib.pyplot as plt
```

```python
# to plot with only one set of value.
```

```python
plt.plot([1,2,3,4,5,6])
plt.xlabel("x label") # show label name
plt.ylabel("y label")
plt.title('first plot') # print name of title
plt.show() 
```

```python
# use  of format argument.....fmt
```

```python
plt.plot([1,2,3,4,5,6],"g") # "g" is frmt argument reperesnt line color.
plt.xlabel("x label") # show label name
plt.ylabel("y label")
plt.title('second plot') # print name of title
plt.show() 
```

```python
# key argument . 
```

```python
plt.plot([1,2,3,4,5,6],"b+" , markersize=15,) # "g" is frmt argument reperesnt line color and "+" is marker size..Marker size iskey argument.
plt.xlabel("x label") # show label name
plt.ylabel("y label")
plt.title('second plot') # print name of title
plt.show() 
```

```python
plt.plot([1,2,3,4,5,6],"r" , linewidth=15,)  #key argument used for line.
plt.xlabel("x label")
plt.ylabel("y label")
plt.title('second plot') 
plt.show() 
```

```python
# Basic formnatting using key word  argument and format argument.
```

```python
plt.plot([1,2,3,4,5,6],"g" , color="red",) .
plt.xlabel("x label") 
plt.ylabel("y label")
plt.title('second plot') 
plt.show() 
```

```python
plt.plot([1,2,3,4,5,6],"r"  )
plt.plot([2,4,5,7,8,9],"bo"  )
plt.xlabel("x label") 
plt.ylabel("y label")
plt.title('second plot') 
plt.show() 
```

```python
#To represent both set in one line
```

```python
plt.plot([1,2,3,4,5,6],"r" ,[2,4,5,7,8,9],"b-."  ) 
plt.xlabel("x label") 
plt.ylabel("y label")
plt.title('second plot') 
plt.show()
```
