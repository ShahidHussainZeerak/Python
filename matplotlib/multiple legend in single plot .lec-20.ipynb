```python
import  matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4,5])
l1=plt.plot(a**2,a,color="red")
l2=plt.plot(a**3,a,color='green')
l3=plt.plot(a,a**2,color="black")
l4=plt.plot(a,a**3,color='orange')
plt.xlabel("x-label",)
plt.ylabel("y-label")
plt.title("First plot")
legend1=plt.legend(["l1"],loc="upper left")
plt.gca().add_artist(legend1)
legend2=plt.legend(["l2"],loc="upper right")
plt.gca().add_artist(legend2)
legend1=plt.legend(["l3","l4"],loc="center")
plt.tight_layout()
plt.show()
```
